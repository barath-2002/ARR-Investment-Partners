### Check Exchange Rate Ranges

This cell defines a function `check_exchange_rate_ranges` to validate whether the exchange rates for different currencies fall within predefined reasonable ranges. It flags rows where the exchange rate is outside the expected range and returns a DataFrame containing the flagged rows. 

#### Formula:
$$
\text{Exchange Rate Error} = 
\begin{cases} 
\text{True}, & \text{if } \text{Exchange Rate} \notin [\text{Min Range}, \text{Max Range}] \\
\text{False}, & \text{otherwise}
\end{cases}
$$


```python
def check_exchange_rate_ranges(data):
    currency_ranges = {
        'USD': (1, 1),
        'EURUSD': (0.95, 1.2),
        'GBPUSD': (1.20, 1.45),
        'HKDUSD': (0.127, 0.129),
        'CADUSD': (0.69, 0.83),
        'SEKUSD': (0.087, 0.125),
        'CHFUSD': (0.91, 1.18),
        'JPYUSD': (0.0063, 0.0111),
        'PLNUSD': (0.22, 0.31)
    }
    
    data['Exchange Rate Error'] = data.apply(
        lambda row: not (currency_ranges[row['Currency']][0] <= row['Exchange Rate'] <= currency_ranges[row['Currency']][1])
        if row['Currency'] in currency_ranges else False,
        axis=1
    )
    exchange_rate_errors = data[data['Exchange Rate Error']]
    
    return exchange_rate_errors[['Date', 'P_Ticker', 'Currency', 'Exchange Rate', 'Exchange Rate Error']]

# Example usage
exchange_rate_errors = check_exchange_rate_ranges(data)
exchange_rate_errors
```

### Check for quantity change errors

This function, `check_quantity_trade_mismatch`, identifies discrepancies between the change in holding quantities (`Open Quantity` and `Close Quantity`) and the reported trades (`Traded Today`). It calculates the `Quantity Change` as the difference between `Close Quantity` and `Open Quantity`. Rows where the `Quantity Change` is non-zero but does not match the `Traded Today` value (within a small numerical tolerance, `atol=1e-6`) are flagged.

#### Formula:
$$
\text{Quantity Change} = \text{Close Quantity} - \text{Open Quantity}
$$

```python
def check_quantity_trade_mismatch(data):
    data['Quantity Change'] = data['Close Quantity'] - data['Open Quantity']
    quantity_flags = data[
        (data['Quantity Change'] != 0) & 
        (~np.isclose(data['Traded Today'], data['Quantity Change'], atol=1e-6))
    ]
    return quantity_flags[['Date', 'P_Ticker', 'Open Quantity', 'Close Quantity', 'Quantity Change', 'Traded Today']]

# Example usage
quantity_flags = check_quantity_trade_mismatch(data)
quantity_flags
```

### Check Prices where changes are greater than 10 percent compared to yesterday's price

This cell defines a function `check_inconsistent_prices` to identify rows in the dataset where the percentage change in price exceeds a specified threshold (10% in this case). It calculates the price change and the percentage change relative to the previous day's price, flags rows with significant changes, and returns a DataFrame containing the flagged rows.

#### Formula:
$$
\text{Price Change Percentage} = \frac{\text{Price} - \text{Price Yesterday}}{\text{Price Yesterday}}
$$


```python
def check_inconsistent_prices(data):
    data['Price Change'] = data['Price'] - data['Price Yesterday']
    data['Price Change Percentage'] = data['Price Change'] / data['Price Yesterday']
    price_flags = data[np.abs(data['Price Change Percentage']) > 0.1]  # Flagging >10% change
    return price_flags[['Date', 'P_Ticker', 'Price', 'Price Yesterday', 'Price Change Percentage']]

# Example usage
price_flags = check_inconsistent_prices(data)
price_flags
```

### Calculate and Check Profit Errors

This cell defines a function `calculate_and_check_profit_errors` to calculate the Dollar Profit and Loss (PnL) based on each row of data and including if a trade has been carried out and compare it with the existing Dollar PnL values. It flags rows where the calculated Dollar PnL does not match the existing Dollar PnL within a specified tolerance.

#### Formula:
1. **Dollar PnL Calculation**:
    - If `Open Quantity` equals `Close Quantity`:
      $$
      \text{Dollar PnL} = (\text{Price} - \text{Price Yesterday}) \times \text{Close Quantity} \times \text{Exchange Rate}
      $$
    - If `Close Quantity` is greater than `Open Quantity`:
      $$
      \text{Dollar PnL} = \text{Open Quantity} \times (\text{Price} - \text{Price Yesterday}) \times \text{Exchange Rate} - (\text{Close Quantity} - \text{Open Quantity}) \times (\text{Trade Price} - \text{Price}) \times \text{Exchange Rate}
      $$
    - If `Close Quantity` is less than `Open Quantity`:
      $$
      \text{Dollar PnL} = (\text{Open Quantity} - \text{Close Quantity}) \times (\text{Trade Price} - \text{Price}) \times \text{Exchange Rate} + \text{Open Quantity} \times (\text{Price} - \text{Price Yesterday}) \times \text{Exchange Rate}
      $$

```python
def calculate_and_check_profit_errors(data):
    # Calculate Dollar PnL
    data['Calculated Dollar PnL'] = np.where(
        data['Open Quantity'] == data['Close Quantity'],
        (data['Price'] - data['Price Yesterday']) * data['Close Quantity'] * data['Exchange Rate'],
        np.where(
            data['Close Quantity'] > data['Open Quantity'],
            data['Open Quantity'] * (data['Price'] - data['Price Yesterday']) * data['Exchange Rate'] - 
            (data['Close Quantity'] - data['Open Quantity']) * (data['Trade Price'] - data['Price']) * data['Exchange Rate'],
            (data['Open Quantity'] - data['Close Quantity']) * (data['Trade Price'] - data['Price']) * data['Exchange Rate'] +
            data['Open Quantity'] * (data['Price'] - data['Price Yesterday']) * data['Exchange Rate']
        )
    )

    hidden_flags = data[
        ~np.isclose(data['Dollar PnL'], data['Calculated Dollar PnL'], atol=1e-6)
    ]
    
    return hidden_flags[['Date', 'P_Ticker', 'Dollar PnL', 'Calculated Dollar PnL']]

# Example usage
pnl_flags = calculate_and_check_profit_errors(data)
pnl_flags
```

### Calculate Value in USD and Exchange Rate Errors

This function, `check_currency_conversion_errors`, identifies discrepancies in currency conversion calculations within a dataset. It computes the "Calculated Value in USD" for each row by multiplying the `Price`, `Exchange Rate`, and `Close Quantity`. It then compares this calculated value with the existing `Value in USD` column using a tolerance (`atol=1e-6`) to account for minor numerical differences.

#### Formula:
$$
\text{Calculated Value in USD} = \text{Price} \times \text{Exchange Rate} \times \text{Close Quantity}
$$

```python
def check_currency_conversion_errors(data):
    data['Calculated Value in USD (Currency Check)'] = data['Price'] * data['Exchange Rate'] * data['Close Quantity']
    data['Currency Conversion Error'] = ~np.isclose(data['Value in USD'], data['Calculated Value in USD (Currency Check)'], atol=1e-6)
    
    # Filter rows with currency conversion errors
    currency_conversion_errors = data[data['Currency Conversion Error']]
    return currency_conversion_errors[['Date', 'P_Ticker', 'Value in USD', 'Calculated Value in USD (Currency Check)']]

currency_conversion_errors = check_currency_conversion_errors(data)
currency_conversion_errors
```

### Validate Opening and Closing Weights

Given the identified discrepancies in the `Value in USD`, it is crucial to validate the accuracy of the `Opening Weights` and `Closing Weights` as they are directly dependent on the total `Value in USD`.

#### Formulas:
1. **Opening Weights**:  
    $$
    \text{Calculated Opening Weights} = \frac{\text{Price Yesterday} \times \text{Exchange Rate} \times \text{Open Quantity}}{\text{NAV Yesterday}}
    $$

2. **Closing Weights**:  
    $$
    \text{Calculated Closing Weights} = \frac{\text{Price} \times \text{Close Quantity} \times \text{Exchange Rate}}{\text{Calculated NAV}}
    $$

```python
def check_weights_errors(data):
    # Calculate Opening Weights
    data['Calculated Opening Weights'] = (data['Price Yesterday'] * data['Exchange Rate'] * data['Open Quantity']) / data['NAV Yesterday']
    data['Opening Weights Error'] = ~np.isclose(data['Opening Weights'], data['Calculated Opening Weights'], atol=1e-6)

    # Calculate Closing Weights
    data['Calculated Closing Weights'] = (data['Price'] * data['Close Quantity'] * data['Exchange Rate']) / data['Calculated NAV']
    data['Closing Weights Error'] = ~np.isclose(data['Closing Weights'], data['Calculated Closing Weights'], atol=1e-6)

    # Filter rows with errors
    weights_errors = data[data['Opening Weights Error'] | data['Closing Weights Error']]
    return weights_errors[['Date', 'P_Ticker', 'Opening Weights', 'Calculated Opening Weights', 
                           'Opening Weights Error', 'Closing Weights', 'Calculated Closing Weights', 
                           'Closing Weights Error']]

# Example usage
weights_errors = check_weights_errors(data)
weights_errors
```

### Check Market Cap Errors

This function, `check_market_cap_errors`, identifies discrepancies in market capitalization calculations within a dataset. It computes the "Calculated Market Cap" for each row by multiplying the `SharesOut`, `Price`, and `Exchange Rate`. It then compares this calculated value with the existing `Market_Cap` column using a tolerance (`atol=1e-6`) to account for minor numerical differences.

#### Formula:
$$
\text{Calculated Market Cap} = \text{SharesOut} \times \text{Price} \times \text{Exchange Rate}
$$

```python
def check_market_cap_errors(data):
    # Calculate Market Cap
    data['Calculated Market Cap'] = data['SharesOut'] * data['Price'] * data['Exchange Rate']
    data['Market Cap Error'] = ~np.isclose(data['Market_Cap'], data['Calculated Market Cap'], atol=1e-6)
    
    # Filter rows with market cap errors
    market_cap_errors = data[data['Market Cap Error']]
    return market_cap_errors[['Date', 'P_Ticker', 'Market_Cap', 'Calculated Market Cap', 'Market Cap Error']]

# Example usage
market_cap_errors = check_market_cap_errors(data)
market_cap_errors
```

### Check Stock Movement and Performance

This function, `check_stock_movement_and_performance`, recalculates the stock movement and performance contribution for each row in the dataset. It compares the recalculated values with the existing ones to identify discrepancies. The function flags rows where the calculated stock movement or performance contribution does not match the existing values within a specified tolerance (`atol=1e-6`). 

#### Formulas:
1. **Stock Movement**:  
    $$
    \text{Calculated Stock Movement} = \frac{\text{Price} - \text{Price Yesterday}}{\text{Price Yesterday}}
    $$

2. **Performance Contribution**:  
    $$
    \text{Calculated Performance Contribution} = \text{Close Quantity} \times \text{Price} \times \text{Exchange Rate} \times \text{Calculated Stock Movement}
    $$

```python
def check_stock_movement_and_performance(data):
    # Recalculate Stock Movement and Performance Contribution
    data['Calculated Stock Movement'] = (data['Price'] - data['Price Yesterday']) / data['Price Yesterday']
    data['Calculated Performance Contribution'] = data['Close Quantity'] * data['Price'] * data['Exchange Rate'] * data['Calculated Stock Movement']
    
    # Check for discrepancies
    data['Stock Movement Error'] = ~np.isclose(data['Stock Movement'], data['Calculated Stock Movement'], atol=1e-6)
    data['Performance Contribution Error'] = ~np.isclose(data['Performance Contribution'], data['Calculated Performance Contribution'], atol=1e-6)
    
    # Filter rows with errors
    stock_performance_errors = data[
        data['Stock Movement Error'] | data['Performance Contribution Error']
    ]
    
    return stock_performance_errors[['Date', 'P_Ticker', 'Stock Movement', 'Calculated Stock Movement', 
                                     'Stock Movement Error', 'Performance Contribution', 
                                     'Calculated Performance Contribution', 'Performance Contribution Error']]

# Example usage
stock_performance_errors = check_stock_movement_and_performance(data)
stock_performance_errors
```

### Check Trade Day Move and Weights

This function, `check_trade_day_move_and_weights`, recalculates the trade day move, trade weight, and return adjustments for rows where trades occurred (`Traded Today` is not NaN). It compares the recalculated values with the existing ones to identify discrepancies. The function flags rows where the calculated trade day move, trade weight, or return adjustments do not match the existing values within a specified tolerance (`atol=1e-6`).

#### Formulas:
1. **Trade Day Move**:  
    $$
    \text{Calculated Trade Day Move} = \frac{\text{Price} - \text{Trade Price}}{\text{Trade Price}}
    $$

2. **Trade Weight**:  
    $$
    \text{Calculated Trade Weight} = \frac{\text{Traded Today} \times \text{Trade Price} \times \text{Exchange Rate}}{\text{NAV Yesterday}}
    $$

3. **Return Adjustments**:  
    $$
    \text{Calculated Return Adjustments} = \text{Calculated Trade Weight} \times \text{Calculated Trade Day Move}
    $$

```python
def check_trade_day_move_and_weights(data):
    # Perform calculations only for rows where 'Traded Today' is not NaN
    data.loc[data['Traded Today'].notna(), 'Calculated Trade Day Move'] = (
        (data['Price'] - data['Trade Price']) / data['Trade Price']
    )
    data.loc[data['Traded Today'].notna(), 'Calculated Trade Weight'] = (
        (data['Traded Today'] * data['Trade Price'] * data['Exchange Rate']) / data['NAV Yesterday']
    )
    data.loc[data['Traded Today'].notna(), 'Calculated Return Adjustments'] = (
        data['Calculated Trade Weight'] * data['Calculated Trade Day Move']
    )
    
    # Check for discrepancies
    data['Trade Day Move Error'] = ~np.isclose(data['Trade Day Move'], data['Calculated Trade Day Move'], atol=1e-6)
    data['Trade Weight Error'] = ~np.isclose(data['Trade Weight'], data['Calculated Trade Weight'], atol=1e-6)
    data['Return Adjustment Error'] = ~np.isclose(data['Return Adjustments'], data['Calculated Return Adjustments'], atol=1e-6)
    
    # Filter rows with errors and where 'Traded Today' is not NaN
    trade_day_move_and_weights_errors = data.loc[
        data['Traded Today'].notna() & 
        (data['Trade Day Move Error'] | data['Trade Weight Error'] | data['Return Adjustment Error'])
    ]
    
    return trade_day_move_and_weights_errors[['Date', 'P_Ticker', 'Trade Day Move', 'Calculated Trade Day Move', 
                                               'Trade Weight', 'Calculated Trade Weight', 
                                               'Return Adjustments', 
                                              'Calculated Return Adjustments']]

# Example usage
trade_day_move_and_weights_errors = check_trade_day_move_and_weights(data)
trade_day_move_and_weights_errors
```
